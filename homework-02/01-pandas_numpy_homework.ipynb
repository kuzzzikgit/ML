13) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.

```python
arrM = np.arange(9).reshape((3,3))
arrMin = arrM.min(axis=1, keepdims=True)
arrMax = arrM.max(axis=1, keepdims=True)
print(arrM)
print(arrMin)
print(arrMax)
arrM1 = np.column_stack((arrM, arrMin[:, 0]))
arrM2 = np.column_stack((arrM1, arrMax[:, 0]))
print(arrM1)
print(arrM2)
```

14) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу эётого произведения.

```python
#До 51 включительно? Если нет , то
array = np.arange(1, 51)
print(array)
arrayOdd = array[array % 2 != 0]
print(arrayOdd)
arrayEven = array[array % 2 == 0]
print(arrayEven)
arrROdd = arrayOdd.reshape((5, 5))
print(arrROdd)
arrREven = arrayEven.reshape((5, 5))
print(arrREven)
arrMult = np.dot(arrROdd, arrREven)
print(arrMult)
arrInv = np.linalg.inv(arrMult)
print(arrInv)
```
